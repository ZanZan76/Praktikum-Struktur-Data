```python
# Queue dengan Collections.deque
from collections import deque

queue = deque()

queue.append('Memet')
queue.append('Joko')
queue.append('Siti')
queue.append('Rusdi')
queue.append('Mukyo')
queue.append('Supriyadi')
queue.append('Mulyono')
queue.append('Dewi')
print('Daftar Antrian : ', queue)

queue.popleft()
queue.popleft()
queue.popleft()
print('Nama Yang Keluar :', queue)

print('Sisa Antrian :', queue)
```

```python
# Queue dengan queue.queue

from queue import Queue

Qu = Queue(maxsize=8)

Qu.put('Rosidi')
Qu.put('Mariadi')
Qu.put('Thiago')
Qu.put('Mari')
Qu.put('Mario')
Qu.put('Maria')
Qu.put('Ratu')
Qu.put('Tyr')
print('Daftar Antrian : ', Qu.queue)
print('Jumlah : ', Qu.qsize())
print('Antrian Penuh : ', Qu.full())

Qu.get()
Qu.get()
Qu.get()
Qu.get()
print('\nNama Yang Keluar : ', Qu.queue)
print('\nSisa Antrian : ', Qu.queue)
print('Jumlah : ', Qu.qsize())
print('Antrian Penuh : ', Qu.full())
```

```python
# Queue dengan Single Linked List
class Queue:
    def __init__(self):
        self.queue = list()
    
    def AddData(self, data):
        if data not in self.queue:
            self.queue.append(data)
            return True
        return False
    
    def RemoveData(self):
        if len(self.queue)>0:
            return self.queue.pop()
        return('Antrian Kosong')
    
    def Size(self):
        return len(self.queue)
    
Q = Queue()
Q.AddData('Mamat')
Q.AddData('Anwar')
Q.AddData('Mariadi')
Q.AddData('Redi')
Q.AddData('Ratu')
Q.AddData('Udin')
Q.AddData('Misa')
print('Jumlah Antrian : ', Q.Size())
print('Daftar Antrian : ', Q.queue)

Q.RemoveData()
Q.RemoveData()
Q.RemoveData()
print('Jumlah Antrian : ', Q.Size())
print('Daftar Antrian : ', Q.queue)
```

```python
# Queue dengan Single Linked List
class Queue:
    def __init__(self, data):
        self.data= data
        self.next = None


class LinkedQueue:
    def __init__(self):
        self.queue = None
        self.next = None
        self.size = 0

    def TampilkanSemua(self):
        hasil = []
        now = self.queue

        while now is not None:
            hasil.appernd(now.data)
            now = now.next
        return hasil
    def AddData(self, data):
        new_data = Queue(data)
        if self.next is None:
            self.queue = self.next = new_data 
        else:
            self.next.next = new_data
            self.next = new_data
        self.size += 1
    
    def RemoveData(self):
        if self.queue is None:
            return('Antrean Kosong')

        removed_data = self.queue.data
        self.queue = self.queue.next

        if self.queue is None:
            self.next = None
        
        self.size -= 1
        return removed_data
    def Size(self):
        return self.size
    
Q = LinkedQueue()
Q.AddData('Mamat')
Q.AddData('Anwar')
Q.AddData('Mariadi')
Q.AddData('Redi')
Q.AddData('Ratu')
Q.AddData('Udin')
Q.AddData('Misa')
print('Jumlah Antrian : ', Q.Size())
print('Daftar Antrian : ', Q.queue.data)

Q.RemoveData()
Q.RemoveData()
Q.RemoveData()
print('Jumlah Antrian : ', Q.Size())
print('Daftar Antrian : ', Q.queue.data)
```
