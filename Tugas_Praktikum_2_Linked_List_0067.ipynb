Double Linked List

```python


class playlist:
    def __init__(self, music):
        self.data = music
        self.next = None
        self.prev = None

class music:
    def __init__(self, judul):
      self.judul = judul

    def __str__(self):
      return self.judul

m1 = music("Emptiness Machine")
m2 = music("A Line Without A Hook")
m3 = music("Let Down")


class home:
    def __init__(self):
        self.head = None
        self.current = None

    def kunjungi(self, music):
        baru = playlist(music)
        print(f"---- Mengunjungi: {music} ----")

        if self.head is None:
            self.head = baru
            self.current = baru
        else:
           if self.current.next:
              self.current.next.prev = None
              self.current.next = None

           self.current.next = baru
           baru.prev = self.current
           self.current = baru
        return baru

    def kunjungi(self, music):
      baru = playlist(music)
      print(f"---- Mengunjungi: {music} ----")

      if self.head is None:
          self.head = baru
          self.current = baru
      else:

          if self.current.next:
              self.current.next.prev = None
              self.current.next = None

          self.current.next = baru
          baru.prev = self.current
          self.current = baru
      return baru

    def back(self):
       if self.current and self.current.prev:
           self.current = self.current.prev
           print(f"Mundur ke: {self.current.data}")
       else:
           print("Tidak ada riwayat pemutaran sebelumnya!")

    # Penyisipan di Tengah
    def AddInBetween(self, middle_song, newsong):
        if middle_song is None:
          print("Lagunya belum ada!")
          return

        new_music = music(newsong)
        Newest = playlist(newsong)

        Newest.next = middle_song.next
        Newest.prev = middle_song
        if middle_song.next is not None:
          middle_song.next.prev = Newest

        middle_song.next = Newest
        print(f"---Berhasil menyisipkan: {newsong}---")

    # Hapus dengan kata kunci
    def RemoveNode(self, Removesong):
        temp = self.head
        target_music = str(Removesong)

        while temp is not None:
          if str(temp.data) == target_music:
            break
          temp = temp.next

        if temp is None:
          print(f"---Lagu '{target_music}' tidak ada dalam playlist-mu---")
          return

        if temp == self.head:
          self.head = temp.next

        if temp.prev is not None:
          temp.prev.next = temp.next

        if temp.next is not None:
          temp.next.prev = temp.prev

        if temp == self.current:
          if temp.prev:
              self.current = temp.prev
          else:
            self.current = temp.next

        print(f"---Lagu Berhasil Di Hapus: {Removesong}---")
        del temp

    def forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"Maju ke: {self.current.data}")
        else:
            print("Sudah di lagu paling depan!")


H = home()
print("\nSebelum dilakukan remove")
m1_music = H.kunjungi(m1)
H.AddInBetween(m1_music, "Guren")
H.kunjungi(m2)
H.kunjungi(m3)

# print("\nSebelum dilakukan remove")
# H.back()

H.RemoveNode(m2)

print("\nSetelah dilakukan remove")
H.back()
H.forward()
```

Circular Linked List

```python
class playlist:
    def __init__(self, music):
        self.music = music
        self.next = None
        self.prev = None


class music:
    def __init__(self, judul):
        self.judul = judul
    def __str__(self):
       return self.judul


class Cycle:
    def __init__(self):
        self.head = None
        self.current_song = None

    # Menyisipkan di akhir
    def AddInEnd(self, newsong_):
      new_music = music(newsong)
      newest = playlist(newsong)

      if self.head is None:
        self.head = newest
        newest.next = newest
        newest.prev = newest
        self.current = newest
      else:
        tail = self.head.prev

      tail.next = newest
      newest.prev = tail

      newest.next = self.head
      self.head.prev = newest

      print(f"Berhasil menyisipkan: {newsong}")

    # Hapus di akhir
    def RemoveEnd(self, Removesong):
        if self.head is None:
          print("Playlist kosong!")
          return

        tail = self.head.prev
        print(f"--- Lagu {tail.music} Berhasil dihapus dari akhir ---")

        while True:
          if str(self.music) == Removesong:
            found = True
            break
          temp = temp.next
          if temp == self.head:
            break

        if not found:
          print(f"---Lagu '{Removesong}' tidak ada dalam playlist-mu---")
          return

        if temp.next == temp:
          self.head = None
          self.current = None
        else:
          temp.prev.next = temp.next
          temp.next.prev = temp.prev

          if temp == self.head:
            self.head = temp.next

          if temp == self.head:
            self.current = temp.next


        print(f"---Lagu Berhasil Di Hapus---")
        del temp


    def kunjungi(self, music):
        baru = playlist(music)

        #print(f"---- Mengunjungi: {music} ----")

        if not self.current_song:
            self.current_song = baru
            baru.next = self.current_song
            baru.prev = self.current_song
        else:
           last = self.current_song.prev

           last.next = baru
           baru.prev = last
           baru.next = self.current_song
           self.current_song.prev = baru
        print(f"Music {music} diputar!")
        return baru

    def next(self):
       if self.current_song:
           self.current_song = self.current_song.prev
           print(f"Mundur ke: {self.current_song.music}")

    def prev(self):
        if self.current_song:
            self.current_song = self.current_song.next
            print(f"Musik yang diputar sekarang: {self.current_song.music}")



m1 = music("Emptiness Machine")
m2 = music("A Line Without A Hook")
m3 = music("Let Down")
m4 = music("Memories")

C = Cycle()


print("Sebelum dihapus")
C.kunjungi(m1)
C.kunjungi(m2)
C.kunjungi(m3)
C.kunjungi(m4)



C.RemoveEnd()
C.AddInEnd("American Idiot")

print("\n --- Setelah dihapus ---")
print(f"Musik yang diputar sekarang: {C.current_song.music}")

C.prev()
C.next()
C.next()
C.next()
```
