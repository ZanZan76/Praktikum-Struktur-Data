Stack dengan List

```python
stack = []

stack.append(1)
stack.append(3)
stack.append(5)
stack.append(6)
stack.append(9)
stack.append(4)

print('Stack :')
print(f'Sebelum dikeluarkan = {stack}')

print('\nYang dikeluarkan:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print(f'\nSetelah dikeluarkan = {stack}')
```

Stack dengan collections deque

```python
from collections import deque


stack = deque()

stack.append(1)
stack.append(3)
stack.append(5)
stack.append(6)
stack.append(9)
# stack.append(4)

print('Stack :')
print(f'Sebelum dikeluarkan = {stack}')

print('\nYang dikeluarkan:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print(f'\nSetelah dikeluarkan = {stack}')






```

Stack dengan queue.LifoQueue

```python
from queue import LifoQueue


stack = LifoQueue(maxsize=6)

stack.put(1)
stack.put(3)
stack.put(5)
stack.put(6)
stack.put(9)
stack.put(4)
# stack.put(2)

print('Full :', stack.full())
print('Size :', stack.qsize())

print('\nYang dikeluarkan:')
print(stack.get())
print(stack.get())
print(stack.get())

print('\nEmpty :', stack.empty())
```

Stack dengan Single Linked List

```python
class Numbers:
  def __init__(self, angka):
   self.angka = angka
   self.next = None

class Stack:
  def __init__(self):
    self.head = Numbers("head")
    self.size = 0

  def __str__(self):
    num = self.head.next
    out = ""
    while num:
      out += str(num.angka) + "->"
      num = num.next
    return out[:-2]

  def getSize(self):
    return self.size

  def isEmpty(self):
    return self.size == 0

  def peek(self):
    if self.isEmpty():
      raise Exception("Peeking from an empty stack")
    return self.head.next.angka

  def push(self, angka):
    numbers = Numbers(angka)
    numbers.next = self.head.next
    self.head.next = numbers
    self.size += 1

  def pop(self):
    if self.isEmpty():
      raise Exception("Popping from an empty stack")
    hapus = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return hapus.angka

if __name__ == "__main__":
  stack = Stack()
  for i in range(1, 11):
    stack.push(i)
  print(f"Stack: {stack}")

  for _ in range(1,6):
    hapus = stack.pop()
    print(f"Pop: {hapus}")
  print(f"Stack: {stack}")















```
